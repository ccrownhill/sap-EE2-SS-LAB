```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def X(omega):
    value = 1 / (1 - 0.8 * np.exp(-1j * omega))
    return np.sqrt(value.real**2 + value.imag**2)

def magX(omega):
    return 1 / np.sqrt(1 - 2 * 0.8 * np.cos(omega) + 0.8**2)

def phaseX(omega):
    return - np.arctan(0.8 * np.sin(omega) / (1 - 0.8 * np.cos(omega)))
```

```python
omega = np.linspace(0, 2 * np.pi, 100)
Xout = magX(omega)
plt.plot(omega, magX(omega))
plt.show()
plt.plot(omega, phaseX(omega))
plt.show()
```
